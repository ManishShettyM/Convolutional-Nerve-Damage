# No GPU ? Not enough RAM ? ... no problem 
 
The dataset has been segregated into 5 batches and packed up using the pickle module in python.
1 batch image = 10000 x 3072 numpy array
10000 cols = no of data items
1 row = 3072 should represent a 32x32 pixel image

 In order to feed an image data into a CNN model, the dimension of the input tensor should be either (width x height x num_channel) or (num_channel x width x height).So some formatting is required.

### Transformations and Reshaping

1. divide the row by 3 parts 
2. divide each of those pieces by 32

Now it is in (num_channel, width, height) form.
But we need  (width, height, num_channel) form , so just transpose the matrix


```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

def load_cifar10_batch(path,batch_id):
     with open(path + '/data_batch_' + str(batch_id), mode='rb') as file:
            batch = pickle.load(file,encoding='latin1')
            
     features = batch['data'].reshape((len(batch['data']),3,32,32)).transpose(0,2,3,1)
     labels = batch['labels']
    
     return features,labels        
```

#### Labels :

* airplane : 0
* automobile : 1
* bird : 2
* cat : 3
* deer : 4
* dog : 5
* frog : 6
* horse : 7
* ship : 8
* truck : 9

```python
def load_label_names():
    return ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

### Exploratory Data Analysis

```python
def display_stats(cifar10_dataset_folder_path, batch_id, sample_id):
    features, labels = load_cifar10_batch(cifar10_dataset_folder_path, batch_id)
    
    if not (0 <= sample_id < len(features)):
        print('{} samples in batch {}.  {} is out of range.'.format(len(features), batch_id, sample_id))
        return None

    print('\nStats of batch #{}:'.format(batch_id))
    print('# of Samples: {}\n'.format(len(features)))
    
    label_names = load_label_names()
    label_counts = dict(zip(*np.unique(labels, return_counts=True)))
    for key, value in label_counts.items():
        print('Label Counts of [{}]({}) : {}'.format(key, label_names[key].upper(), value))
    
    sample_image = features[sample_id]
    sample_label = labels[sample_id]
    
    print('\nExample of Image {}:'.format(sample_id))
    print('Image - Min Value: {} Max Value: {}'.format(sample_image.min(), sample_image.max()))
    print('Image - Shape: {}'.format(sample_image.shape))
    print('Label - Label Id: {} Name: {}'.format(sample_label, label_names[sample_label]))
    
    plt.imshow(sample_image)
```

```python
# say we want to explore the batch3 and sample number 6500
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

batch_id = 3
sample_id = 6500
display_stats("dataset", batch_id, sample_id)
```

```python
batch_id = 2
sample_id = 6400
display_stats("dataset", batch_id, sample_id)
```
